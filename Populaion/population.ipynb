```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df1=df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
df1
```

```python
df_plot = df1.reset_index().melt(id_vars='Continent', 
                                var_name='Year', 
                                value_name='Population')
df_plot
```

```python
df_plot['Year'] = df_plot['Year'].str.extract('(\d+)').astype(int)
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=df_plot,x='Year',y='Population',hue='Continent',marker='o')
plt.title("Continent-wise Population Growth (1970–2022)")
plt.xlabel("Year")
plt.ylabel("Average Population")
plt.legend(title='Continent')
plt.grid(True)

plt.show()
```
